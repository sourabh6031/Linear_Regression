### MAKING A REGRESSOR CLASS

```python
class GDregressor:
    
    def __init__(self,learning_rate = 0.01, epoch=100):
        self.lr = learning_rate
        self.epoch = epoch
        self.coef_ = None
        self.intercept_ = None
    
    def fit(self,X_train,y_train):
        # first taking intercept as 0 and coef_ as 1 for all columns.
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        print(self.coef_)
        
        for i in range(self.epoch):
            y_hat = np.dot(X_train,self.coef_) + self.intercept_   
            intercept_der = -2 * np.mean(y_train - y_hat)
            self.intercept_ = self.intercept_ - (self.lr * intercept_der)
            
            coef_der =  -2 * (np.dot((y_train - y_hat),X_train)/X_train.shape[0])
            self.coef_  = self.coef_ - (self.lr * coef_der)
            
        print(self.coef_, self.intercept_)
            
    
    def predict(self,X_test):
        return np.dot(X_train,self.coef_) + self.intercept_  
```

```python
gdr = GDregressor(epoch=270)
gdr.fit(X_train,y_train)
```

```python
y_pred = gdr.predict(X_test)
y_pred
```
